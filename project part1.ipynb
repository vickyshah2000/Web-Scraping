```python
result, continuation_token = reviews(
    'com.facebook.katana',
    lang='id',
    country='id',
    sort=Sort.MOST_RELEVANT,
    count=500,
    filter_score_with=None)
```

```python
scrapeddata = pd.DataFrame(np.array(result),columns=['review'])

scrapeddata = scrapeddata.join(pd.DataFrame(scrapeddata.pop('review').tolist()))
scrapeddata.head()
```

```python
len(scrapeddata.index)
```

```python
scrapeddata[['content','score','at']].head()
```

```python
scrapeddata1 = scrapeddata[['content','score','at']]
sorteddata = scrapeddata1.sort_values(by='at', ascending=False) #sort by Newest, change to True if you want to sort by Oldest,
sorteddata.head()
```
